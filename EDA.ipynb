```python
#importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('C:/Users/Harshita/Desktop/Stats project Datasets/Unicorn data(funding).csv')
data = pd.read_csv('C:/Users/Harshita/Desktop/Stats project Datasets/full_trak.in_data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
for i in df.index:
  num = ""
  for c in df.iat[i,5]:
    if c.isdigit() or c == '.':
      num = num + c
      df.loc[i, ['Current Valuation']] = [num]
```

```python
df['Current Valuation'] = df['Current Valuation'].astype('float64')
```

```python
df["Funding(Amt in USD)"] = df["Funding(Amt in USD)"].apply(lambda x: str(x).replace(",",""))
df["Funding(Amt in USD)"] = pd.to_numeric(df["Funding(Amt in USD)"],errors='coerce')
```

```python
df1 = df.dropna()
```

## Exploratory Data Analysis

```python
fig = plt.figure(figsize=(8,5))
sns.countplot(x = 'Unicorn Entry Year', data = df)

plt.xlabel("Unicorn Entry Year")
plt.ylabel("Number of startups")
```

```python
fig = plt.figure(figsize=(8,5))
ax = sns.countplot(x = 'Unicorn Entry Year',color='b', data = df)

plt.xlabel("Unicorn Entry Year")
plt.ylabel("Number of startups")

for p in ax.patches:
        ax.annotate('%{:.1f}'.format(p.get_height()), (p.get_x()+0.1, p.get_height()+50))
```

```python
fig = plt.figure(figsize=(15,5))
plt.bar(df['location'], df['Current Valuation'],
        width = 0.4)
 
plt.xlabel("Location")
plt.ylabel("Valuation of the startup")
plt.title("Valuation on location of the startup")
plt.show()
```

```python
fig = plt.figure(figsize=(15,5))
plt.bar(df['location'], df['Funding(Amt in USD)'],
        width = 0.4,color='b')
 
plt.xlabel("Location")
plt.ylabel("Funding")
plt.title("Funding of startups based on location")
plt.show()
```
