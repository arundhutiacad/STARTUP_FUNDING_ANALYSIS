```python
#importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('C:/Users/Harshita/Desktop/Stats project Datasets/GDP (funding).csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df["Number of startups"] = df["Number of startups"].apply(lambda x: str(x).replace(",",""))
df["Number of startups"] = pd.to_numeric(df["Number of startups"],errors='coerce')

df["employment created by startups"] = df["employment created by startups"].apply(lambda x: str(x).replace(",",""))
df["employment created by startups"] = pd.to_numeric(df["employment created by startups"],errors='coerce')
```

```python
df.info()
```

```python
plt.figure(figsize=(8,5))

plt.plot(df['Year'],df['Number of startups'], color='g', label='Number of startups')
plt.plot(df["Year"], df["employment created by startups"],color='b', label='Employment created by startups')

plt.xlabel("Year")
plt.title("Number of Startups VS Employment created by Startups")
plt.legend()
plt.show()
```
